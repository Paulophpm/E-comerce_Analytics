Esse relatório consistem em analisar os dados de consumo de um determinado site com o objetivo de entender o comportamento de compras dos consumidores dessa loja e assim identificar os principais insights que foram gerados a melhoras as vendas da empresa.

Os dados analisados foram obtidos de através de uma estração do próprio site em um arquivos .JSON (Java Script Object Notation), que contém os dados das vendas realizados dentro de um determinado período do ano. Todos os dados de vendas, considarem foi utilizados métodos para deixar cada um dos consumidores anônimos

## Análise de Compras por Gênero

Analisando as 780 vendas efetuadas, observamos que o público masculino correspondeu a 27% (633) dessas transações, enquanto o público feminino representou apenas 5% (136) das vendas. Além disso, 11 (0,4%) pessoas optaram por não divulgar seu gênero. É importante ressaltar que as vendas realizadas pelo público masculino (27%), contribuíram com 81,6% do faturamento da loja, equivalente a R$ 1.867,68. Em contrapartida, as vendas realizadas pelo público feminino (5%) foram responsáveis por 16,7% do faturamento, totalizando R$ 382,91.
